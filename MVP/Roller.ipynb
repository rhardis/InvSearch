```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('regimes_prices.csv')
df = df.iloc[-10:,:]
df = df.reset_index()
df['idx'] = df.index
df.tail()
```

```python
def rf(x, lb=3):
    if lb > (x.idx + 1):
        start = 0
    else:
        start = x.idx + 1 - lb
    ret = np.sum(df[(df['idx'] >= (start)) & (df['idx'] <= x.idx)]['State'])
    return ret
```

```python
df['roll'] = df.apply(lambda x: rf(x, lb=3), axis=1)
df.head(10)
```
